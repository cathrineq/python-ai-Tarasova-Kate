## 1. Что такое функция? Анатомия функции (слайд 3)

Функция — это **блок кода с именем**, который мы:

1. **определяем** один раз с помощью `def`,
2. **вызываем** много раз по имени.

Анатомия простейшей функции:

```python
def introduce():      # def + имя + скобки + двоеточие
    print("Привет!")  # тело функции с отступом

````python
### Ячейка 3 — Простая функция `introduce()` (слайд 5, код)

# 🐱 Функция-представление: кот Байтик рассказывает о себе

def introduce():
    print("=" * 50)
    print("🐱 Привет, я кот Байтик!")
    print("📍 Я студент 2 курса")
    print("🏫 Изучаю Data Science в MIT")
    print("💻 Мой язык: Python")
    print("=" * 50)

# Вызовем функцию несколько раз
introduce()
print("\nЕщё раз представимся 👇")
introduce()
````

## 2. Параметры функции: представляем разных котов

На слайде функция `introduce()` всегда рассказывала про одного и того же кота Байтика.
Чтобы не копировать код и не писать вторую функцию для Фунтика, можно использовать **параметры**:

```python
def introduce_cat(name, course, university):
    ...

````python
# 🐱 Обобщённая функция-представление для любого кота

def introduce_cat(name, course, university, language="Python"):
    print("=" * 50)
    print(f"🐱 Привет, я кот {name}!")
    print(f"📍 Я студент {course} курса")
    print(f"🏫 Учусь в {university}")
    print(f"💻 Мой язык: {language}")
    print("=" * 50)

# Байтик из MIT
introduce_cat("Байтик", 2, "MIT")

# Фунтик из ПетрГУ
introduce_cat("Фунтик", 1, "ПетрГУ", language="Python + AI")

# Попробуйте сами:
# introduce_cat("Мурзик", 3, "СПбГУ", language="R + SQL")
````

## 3. Параметры, кофемашина и глобальная переменная (слайды 4–6)

На слайдах была аналогия:

- функция — это **кофемашина** ☕,
- параметр — это **тип кофе** (“латте”, “эспрессо”, ...).

Пример с Фунтиком:

- глобальная переменная `total_food` хранит, **сколько всего он съел за день**;
- функция `add_food(amount)`:
  - принимает параметр `amount` (сколько граммов кот съел прямо сейчас),
  - добавляет это количество к `total_food`.

Важно понять:

- **параметр** — это *место в функции*, куда подставляется значение;
- **аргумент** — это *конкретное значение*, которое мы передаём при вызове.

````python
# 💰 Фунтик считает, сколько он съел за день

# Глобальная переменная для накопления еды
total_food = 0

def add_food(amount):
    """
    Увеличивает глобальную переменную total_food
    на amount граммов и печатает промежуточный результат.
    """
    global total_food
    total_food = total_food + amount
    print(f"🐱 Фунтик съел {amount} г. Всего: {total_food} г")

# Завтрак
add_food(50)

# Обед
add_food(70)

# Ужин
add_food(60)

print(f"\n📊 Итого за день: {total_food} граммов")
````

## 5. Функция с двумя параметрами: кот Байтик и его биткойн‑кошелёк (слайд 8)

Пример со слайда:

- глобальная переменная `balance` хранит баланс в биткойнах (₿);
- функция `transaction(amount, name)`:
  - принимает **два параметра**:
    - `amount` — сколько пришло или ушло,
    - `name` — название операции;
  - внутри использует `if`, чтобы красиво оформить доход или расход.

Сейчас повторим этот пример и чуть его расширим.

````python
# 💻 Кот Байтик проверяет свой электронный кошелёк

balance = 0  # глобальный баланс в биткойнах

def transaction(amount, name):
    """
    Меняет глобальную переменную balance на amount
    и печатает строку о транзакции.
    """
    global balance
    balance = balance + amount
    if amount > 0:
        print(f"✅ {name}: +{amount}₿ | Баланс: {balance}₿")
    else:
        print(f"❌ {name}: {amount}₿ | Баланс: {balance}₿")

# Поступления и траты
transaction(100, "Зарплата")
transaction(-30, "Корм")
transaction(-20, "Игрушки")
transaction(50, "Премия")

print(f"\n💰 Итоговый баланс: {balance}₿")
````

## 6. Альтернатива `global`: возвращаем результат из функции

В примерах выше мы использовали `global`, чтобы менять переменную `balance` во всей программе.
Иногда удобнее, когда функция **возвращает новое значение**, а мы сами присваиваем его переменной.

Сделаем небольшую версию `transaction_no_global`, которая возвращает обновлённый баланс.

````python
def transaction_no_global(balance, amount, name):
    """
    Не использует global: получает баланс и сумму,
    возвращает новый баланс.
    """
    new_balance = balance + amount
    if amount > 0:
        print(f"✅ {name}: +{amount}₿ | Баланс: {new_balance}₿")
    else:
        print(f"❌ {name}: {amount}₿ | Баланс: {new_balance}₿")
    return new_balance

balance2 = 0
balance2 = transaction_no_global(balance2, 100, "Зарплата")
balance2 = transaction_no_global(balance2, -30, "Корм")
balance2 = transaction_no_global(balance2, -20, "Игрушки")
balance2 = transaction_no_global(balance2, 50, "Премия")

print(f"\n💰 Итоговый баланс (без global): {balance2}₿")
````

## 7. Кот‑шпион Штирлиц: секретный дневник и функция `cat_encode` (слайд 9)

````python
# 🕵 Кот Штирлиц ведёт секретный дневник

def cat_encode(text):
    """
    Заменяет важные слова на 'кошачий код'.
    """
    code = {"встреча": "мяу-мяу", "враг": "гав-гав", "успех": "мур-мур"}
    for word, cat_word in code.items():
        text = text.replace(word, cat_word)
    return text

report = "Сегодня была встреча. Враг не заметил. Успех!"

# Записываем зашифрованный текст в файл
with open("secret_diary.txt", "w", encoding="utf-8") as f:
    f.write(cat_encode(report))

# Читаем текст обратно из файла
with open("secret_diary.txt", "r", encoding="utf-8") as f:
    print("🔒 Секретный дневник:")
    print(f.read())

# Дополнительно: закодируем ещё одно сообщение без записи в файл
extra_report = "Новая встреча завтра. Успех гарантирован."
print("\n📝 Дополнительное зашифрованное сообщение:")
print(cat_encode(extra_report))
````
