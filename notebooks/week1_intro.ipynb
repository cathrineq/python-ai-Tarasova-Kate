## üìñ –ß—Ç–æ –≤ —ç—Ç–æ–º –Ω–æ—É—Ç–±—É–∫–µ?

–ó–¥–µ—Å—å –≤—ã –Ω–∞–π–¥—ë—Ç–µ –ø–µ—Ä–≤—ã–µ –ø—Ä–æ—Å—Ç—ã–µ –ø—Ä–∏–º–µ—Ä—ã –∫–æ–¥–∞ –Ω–∞ Python:
- –£—Å–ª–æ–≤–∏—è (`if`)
- –¶–∏–∫–ª—ã (`for`)
- –†–∞–±–æ—Ç–∞ —Å–æ —Å–ø–∏—Å–∫–∞–º–∏

**–ó–∞–ø—É—Å–∫–∞–π—Ç–µ —è—á–µ–π–∫–∏ –ø–æ –æ—á–µ—Ä–µ–¥–∏** (Shift + Enter) –∏ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä—É–π—Ç–µ —Å –∫–æ–¥–æ–º!

```python
# –ü—Ä–æ—Å—Ç–æ–π –ø—Ä–∏–º–µ—Ä
E = True  # –≠–Ω–µ—Ä–≥–∏—è
if E:
    print("–ß–∞–π–Ω–∏–∫ –≤–∫–ª—é—á—ë–Ω!")

# –°–ª–æ–∂–Ω—ã–π –ø—Ä–∏–º–µ—Ä
E = True
W = True  # –í–æ–¥–∞
if E and W:
    print("–ß–∞–π–Ω–∏–∫ –≤–∫–ª—é—á—ë–Ω –∏ –≤–æ–¥–∞ –Ω–∞–≥—Ä–µ–≤–∞–µ—Ç—Å—è!")
```

```python
# –ì–µ—Ä–æ–∏ —Å–∫–∞–∑–∫–∏ –ø—Ä–æ —Ä–µ–ø–∫—É
heroes = [
    "üëµ –ë–∞–±–∫–∞",
    "üë¥ –î–µ–¥–∫–∞",
    "üë©‚Äçüåæ –í–Ω—É—á–∫–∞",
    "üêï –ñ—É—á–∫–∞",
    "üê± –ö–æ—à–∫–∞",
    "üê≠ –ú—ã—à–∫–∞"
]

# –ù–∞—á–∏–Ω–∞–µ–º —Å –±–∞–±–∫–∏
chain = heroes[0]

# –ü–æ—Å—Ç–µ–ø–µ–Ω–Ω–æ –¥–æ–±–∞–≤–ª—è–µ–º –≥–µ—Ä–æ–µ–≤
for hero in heroes[1:]:
    chain += " ‚Üí " + hero

print(chain)  # –í—ã–≤–æ–¥: "üëµ –ë–∞–±–∫–∞ ‚Üí üë¥ –î–µ–¥–∫–∞ ‚Üí üë©‚Äçüåæ –í–Ω—É—á–∫–∞ ‚Üí üêï –ñ—É—á–∫–∞ ‚Üí üê± –ö–æ—à–∫–∞ ‚Üí üê≠ –ú—ã—à–∫–∞"
```
